# **Text-to-Image Generation using Stable Diffusion and Diffusers**

```python
#diffusers is a hugging face page for using diffusion models from huggingface hub
!pip install diffusers transformers accelerate

# Import StableDiffusionPipeline here
from diffusers import StableDiffusionPipeline
import matplotlib.pyplot as plt
import torch

!pip show torch

```

```python
model_id1 = "dreamlike-art/dreamlike-diffusion-1.0"
model_id2 = "stabilityai/stable-diffusion-xl-base-1.0"

# Now StableDiffusionPipeline is defined and accessible
pipe = StableDiffusionPipeline.from_pretrained(model_id1, torch_dtype=torch.float16, use_safetensors=True)
pipe = pipe.to("cuda")
```

```python
prompt = """A young woman with long black hair styled in two braids stands on stone steps in a mysterious forest shrouded in thick fog. She wears a black dress with a white collar, inspired by Victorian aesthetics. The background features ancient arches covered in creeping vines and glowing lanterns casting a warm light. The lighting is soft and diffused, creating an atmospheric haze with a slightly eerie ambiance. The color palette emphasizes dark and cool tones, with warm highlights from the lanterns. The scene conveys a sense of mystery and intrigue, perfect for a gothic or mystical setting.
"""
```

```python
image = pipe(prompt).images[0]
```

```python
image
```

```python
print("[PROMPT]: ",prompt)
plt.imshow(image);
plt.axis('off');
```

```python
prompt2 = """Santa leaping from a massive cyberpunk skyscraper, his glowing red and green wingsuit lighting up the air around him. Below, a sprawling city pulses with holiday-themed neon signs and glowing snow. Santa releases a shower of high-tech presents mid-dive, each gift deploying tiny jet boosters to reach its destination."""

image = pipe(prompt2).images[0]
```

```python
print('[PROMPT]: ',prompt2)
plt.imshow(image);
plt.axis('off');
```
